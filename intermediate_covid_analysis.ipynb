# LinearRegression


```python
from sklearn.model_selection import train_test_split
```

```python
import datetime as dt
india_cases['date']  = pd.to_datetime(india_cases['date'])
india_cases.head()
```

```python
india_cases['date']=india_cases['date'].map(dt.datetime.toordinal)
india_cases.head()
```

```python
#separate independent and dependent variable
x=india_cases['date']
y=india_cases['total_cases']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
y_pred=model.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
mean_squared_error(x_test,y_pred)
```
